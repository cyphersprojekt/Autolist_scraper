```python
#Asignar una variable a response.json() para que sea más fácil acceder
resultado_json = response.json()
```

```python
#Visualizar los elementos dentro de la clave "search_results" del diccionario resultado_json
#### COMENTADO PARA AHORRAR ESPACIO ####
#resultado_json['search_results']
```

```python
#Asignar una variable a resultado_json['search_results'] para que sea más fácil acceder
search_results = resultado_json['search_results']
```

```python
#Crear listas vacías a las que se agregarán los datos que queremos obtener del website
modelo = []
millaje = [] #Se puede convertir en kilometraje haciendo int() y luego * 1.60934
año = []
concesionaria = []
precio = []
```

```python
#Crear un loop que en orden añada uno por uno los datos que queremos a las listas que definimos en el paso anterior
for resultado in search_results:
    modelo.append(resultado['model'])
    millaje.append(resultado['mileage'])
    año.append(resultado['year'])
    concesionaria.append(resultado['dealer_name'])
    precio.append(resultado['price'])
```

```python
#Crear y visualizar un dataframe para estimar si nuestros resultados son correctos
tesla_df = pd.DataFrame({'Modelo':modelo, 'Millaje':millaje, 'Año':año, 'Concesionaria':concesionaria, 'Precio':precio})
tesla_df
```
